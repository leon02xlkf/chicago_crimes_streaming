```python
from models import crime_deserializer
from kafka import KafkaConsumer
```

```python
server = 'localhost:9092'
topic_name = 'chi_crimes'

consumer = KafkaConsumer(
    topic_name,
    bootstrap_servers=[server],
    auto_offset_reset='earliest',
    group_id='crimes-console',
    value_deserializer=crime_deserializer
)
```

```python
from datetime import datetime

print(f"Listening to {topic_name}...")

count = 0
for message in consumer:
    crime = message.value
    event_dt = datetime.fromtimestamp(crime.event_time / 1000)
    print(f"Crime Type: {crime.crime_type}, Community Area: {crime.community_area}, "
          f"Latitude: {crime.latitude}, Longitude: {crime.longitude}, "
          f"Event Time: {event_dt}")
    count += 1
    if count >= 10:
        print(f"\n... received {count} messages so far (stopping after 10 for demo)")
        break

consumer.close()
```
